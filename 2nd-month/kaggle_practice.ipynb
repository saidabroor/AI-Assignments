```python
df_finance = pd.read_csv(r"C:\Users\Saidabrorkhon\ML_Lectures\ML_Lectures2\Startup Failure (Finance and Insurance).csv", on_bad_lines='skip')
df_food = pd.read_csv(r"C:\Users\Saidabrorkhon\ML_Lectures\ML_Lectures2\Startup Failure (Food and services).csv", on_bad_lines = 'skip')
df_health = pd.read_csv(r"C:\Users\Saidabrorkhon\ML_Lectures\ML_Lectures2\Startup Failure (Health Care).csv", on_bad_lines = 'skip')
df_production = pd.read_csv(r"C:\Users\Saidabrorkhon\ML_Lectures\ML_Lectures2\Startup Failure (Manufactures).csv", on_bad_lines = 'skip')
df_commerce = pd.read_csv(r"C:\Users\Saidabrorkhon\ML_Lectures\ML_Lectures2\Startup Failure (Retail Trade).csv", on_bad_lines = 'skip')
df_media = pd.read_csv(r"C:\Users\Saidabrorkhon\ML_Lectures\ML_Lectures2\Startup Failures (Information Sector).csv", on_bad_lines = 'skip')
```

```python
datasets = {
    "Finance And Insurance": df_finance,
    "Food and Services": df_food,
    "HealthCare": df_health,
    "Manufacturers": df_production,
    "Retail Trade": df_commerce,
    "Information Sector": df_media
}
```

```python
for name, dataset in datasets.items():
  print(f"Information About {name} Dataset: ")
  print('Shape:', dataset.shape)
```

```python
column_sets = {name: set(df.columns) for name, df in datasets.items()}
all_columns = set.union(*column_sets.values())

for name, cols in column_sets.items():
    missing_cols = all_columns - cols
    extra_cols = cols - all_columns
    if missing_cols or extra_cols:
        print(f"Dataset '{name}' has differences:")
        if missing_cols:
            print(f"  Missing columns: {missing_cols}")
        if extra_cols:
            print(f"  Extra columns: {extra_cols}")
```

```python
datasets_columns_same = {
    "Finance And Insurance": df_finance,
    "HealthCare": df_health,
    "Manufacturers": df_production,
    "Retail Trade": df_commerce,
    "Information Sector": df_media
}

columns_list = [set(df.columns) for df in datasets_columns_same.values()]
if all(cols == columns_list[0] for cols in columns_list):
    print("All datasets have the same columns")
    df = pd.concat(datasets.values(), ignore_index=True)
    print("Merged dataset shape:", df.shape)
else:
    print("Datasets do not have the same columns")
```
